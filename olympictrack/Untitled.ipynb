```python
X = df2.drop(['summer_rank'],axis = 1)
y = df2['summer_rank']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train,y_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressordt = DecisionTreeRegressor(random_state=0)
regressordt.fit(X_train,y_train)
```

```python
y_preddt = regressordt.predict(X_test)
y_preddt
```

```python
regressordt.score(X_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressorlin = LinearRegression()
regressorlin.fit(X_train,y_train)
```

```python
y_predlin = regressorlin.predict(X_test)
```

```python
regressorlin.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressorrfr = RandomForestRegressor(n_estimators=100, criterion='mse')
regressorrfr.fit(X_train,y_train)
```

```python
y_predrfr = regressorrfr.predict(X_test)
```

```python
regressorrfr.score(X_test,y_test)
```

```python
from sklearn.svm import SVR
regressorsvr = SVR(kernel = 'rbf', C=1.0)
regressorsvr.fit(X_train,y_train)
```

```python
y_predsvr = regressorsvr.predict(X_test)
```

```python
regressorsvr.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regressorrfr,X_train,y_train,cv=5)
scores,scores.mean
```
