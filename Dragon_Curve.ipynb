```python
import matplotlib.pyplot as plt

def dragon_curve(iterations):
    """Generates the string representation of the dragon curve."""
    if iterations == 0:
        return "F"
    else:
        prev = dragon_curve(iterations - 1)
        return prev + "R" + reverse_invert(prev)

def reverse_invert(s):
    """Reverses and inverts the string, swapping L and R."""
    return "".join(["L" if c == "R" else "R" if c == "L" else c for c in s[::-1]])


def plot_dragon_curve(instructions, step_size=1):
    """Plots the dragon curve based on the instructions."""
    x, y = 0, 0
    angle = 0  # Current direction (0: right, 90: up, 180: left, 270: down)
    x_coords = [0]
    y_coords = [0]

    for move in instructions:
        if move == "F":
            if angle == 0:
                x += step_size
            elif angle == 90:
                y += step_size
            elif angle == 180:
                x -= step_size
            elif angle == 270:
                y -= step_size
            x_coords.append(x)
            y_coords.append(y)
        elif move == "R":
            angle = (angle - 90) % 360
        elif move == "L":
            angle = (angle + 90) % 360

    plt.plot(x_coords, y_coords)
    plt.axis("equal")
    plt.title(f"Dragon Curve (Iterations: {iterations})")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

# Example usage:
iterations = 10  # You can increase this for a more complex curve
instructions = dragon_curve(iterations)
plot_dragon_curve(instructions)
```

```python
import turtle

def dragon_curve(t, length, iterations, sign=1):
    """Recursively draw the Dragon Curve using turtle."""
    if iterations == 0:
        t.forward(length)
    else:
        t.right(45 * sign)
        dragon_curve(t, length, iterations - 1, 1)
        t.left(90 * sign)
        dragon_curve(t, length, iterations - 1, -1)
        t.right(45 * sign)

# Setup turtle
screen = turtle.Screen()
screen.bgcolor("white")
t = turtle.Turtle()
t.speed(0)
t.penup()
t.goto(-200, 0)
t.pendown()

# Parameters
iterations = 10      # Increase for more complexity (but slower)
step_length = 8      # Smaller values yield better shapes

# Draw the curve
dragon_curve(t, step_length, iterations)

# Finish
turtle.done()
```
