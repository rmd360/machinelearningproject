```python

import numpy as np

# Coefficients for the quadratic equation ax^2 + bx + c = 0
a = 1
b = -3
c = 2

# Calculate the discriminant
discriminant = b**2 - 4*a*c

# Find the roots
if discriminant > 0:
    root1 = (-b + np.sqrt(discriminant)) / (2*a)
    root2 = (-b - np.sqrt(discriminant)) / (2*a)
    print(f"The roots are {root1} and {root2}")
elif discriminant == 0:
    root = -b / (2*a)
    print(f"The root is {root}")
else:
    real_part = -b / (2*a)
    imaginary_part = np.sqrt(-discriminant) / (2*a)
    print(f"The roots are complex: {real_part} ± {imaginary_part}i")
```

```python

# BFS implementation using a queue
def bfs(graph, start):
    visited = set()
    queue = [start]
    visited.add(start)

    while queue:
        vertex = queue.pop(0)
        print(vertex, end=" ")

        for neighbor in graph[vertex]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.add(neighbor)

# Representing a graph as an adjacency list
graph = {
    1: [2, 3],
    2: [1, 4],
    3: [1],
    4: [2]
}

# Perform BFS starting from node 1
bfs(graph, 1)
```

```python

import collections

def bfs(graph, start):
    visited = set()
    queue = collections.deque([start])
    visited.add(start)

    while queue:
        vertex = queue.popleft()
        print(vertex, end=" ")

        for neighbor in graph[vertex]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.add(neighbor)

# Example graph as adjacency list
graph = {
    1: [2, 3],
    2: [1, 4],
    3: [1],
    4: [2]
}

bfs(graph, 1)  # Output: 1 2 3 4
```

```python

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=" ")

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Example graph as adjacency list
graph = {
    1: [2, 3],
    2: [1, 4],
    3: [1],
    4: [2]
}

dfs(graph, 1)  # Output: 1 2 4 3
```

```python

import heapq

def dijkstra(graph, start):
    # Initialize the distances
    distances = {vertex: float('infinity') for vertex in graph}
    distances[start] = 0
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_vertex = heapq.heappop(priority_queue)

        # Skip processing if the node has already been visited
        if current_distance > distances[current_vertex]:
            continue

        # Process each neighbor of the current node
        for neighbor, weight in graph[current_vertex]:
            distance = current_distance + weight

            # Only consider this new path if it's better
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances

# Example weighted graph (adjacency list with weights)
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('A', 1), ('C', 2), ('D', 5)],
    'C': [('A', 4), ('B', 2), ('D', 1)],
    'D': [('B', 5), ('C', 1)]
}

print(dijkstra(graph, 'A'))  # Shortest paths from 'A'
```

```python

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(gcd(48, 18))  # Output: 6
```

```python

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(29))  # Output: True
print(is_prime(30))  # Output: False
```

```python

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(factorial(5))  # Output: 120
```

```python

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Example usage:
arr = [12, 11, 13, 5, 6, 7]
merge_sort(arr)
print(arr)  # Output: [5, 6, 7, 11, 12, 13]
```

```python

import networkx as nx
import matplotlib.pyplot as plt

edges = [
    ("Alice", "Bob"),
    ("Bob", "Charlie"),
    ("Charlie", "Diana"),
    ("Diana", "Alice"),
    ("Alice", "Eve"),
]

G = nx.Graph()
G.add_edges_from(edges)

# Draw graph
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, node_color='skyblue', edge_color='gray', node_size=2000, font_size=15)
plt.title("User Connection Graph")
plt.show()

# Shortest path
print("Shortest path from Alice to Diana:", nx.shortest_path(G, "Alice", "Diana"))

# Degree of each node
print("Degrees:", dict(G.degree()))
```

```python
relation = [(1, 1), (2, 2), (3, 3), (1, 2), (2, 1), (2, 3), (3, 2)]
elements = {1, 2, 3}

# Reflexive
is_reflexive = all((a, a) in relation for a in elements)
# Symmetric
is_symmetric = all((b, a) in relation for (a, b) in relation)
# Transitive
is_transitive = all((a, c) in relation for (a, b1) in relation for (b2, c) in relation if b1 == b2)

print("Reflexive:", is_reflexive)
print("Symmetric:", is_symmetric)
print("Transitive:", is_transitive)
```

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1])
y = np.array([0, 1])
X, Y = np.meshgrid(x, y)
Z = np.logical_or(np.logical_and(X, Y), np.logical_and(np.logical_not(X), Y)).astype(int)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('F = (A AND B) OR (NOT A AND B)')
plt.title('Boolean Surface')
plt.show()
```

```python

def fib(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series

f = fib(15)
plt.plot(f, marker='o', linestyle='-', color='purple')
plt.title("Fibonacci Sequence")
plt.xlabel("n")
plt.ylabel("F(n)")
plt.grid(True)
plt.show()
```
