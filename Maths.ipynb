```python

import matplotlib.pyplot as plt

# Create data for a line y = 2x + 1
x = np.linspace(-10, 10, 100)
y = 2 * x + 1

# Plot the graph
plt.plot(x, y, label="y = 2x + 1")
plt.title("Linear Equation Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python

import numpy as np

# Coefficients for the quadratic equation ax^2 + bx + c = 0
a = 1
b = -3
c = 2

# Calculate the discriminant
discriminant = b**2 - 4*a*c

# Find the roots
if discriminant > 0:
    root1 = (-b + np.sqrt(discriminant)) / (2*a)
    root2 = (-b - np.sqrt(discriminant)) / (2*a)
    print(f"The roots are {root1} and {root2}")
elif discriminant == 0:
    root = -b / (2*a)
    print(f"The root is {root}")
else:
    real_part = -b / (2*a)
    imaginary_part = np.sqrt(-discriminant) / (2*a)
    print(f"The roots are complex: {real_part} ± {imaginary_part}i")
```

```python

# Parabola equation y = x^2 - 4x + 3
x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 3

plt.plot(x, y, label="y = x^2 - 4x + 3")
plt.title("Parabola")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Define two sets
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# Set operations
union = A.union(B)  # A ∪ B
intersection = A.intersection(B)  # A ∩ B
difference = A.difference(B)  # A - B
symmetric_difference = A.symmetric_difference(B)  # A △ B

print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Difference (A - B): {difference}")
print(f"Symmetric Difference: {symmetric_difference}")
```

```python

import math

n = 5  # number of objects
permutations = math.factorial(n)
print(f"The number of permutations of {n} objects is: {permutations}")
```

```python

from math import comb

n = 5  # total objects
r = 3  # selected objects
combinations = comb(n, r)
print(f"The number of ways to choose {r} objects from {n} objects is: {combinations}")
```
