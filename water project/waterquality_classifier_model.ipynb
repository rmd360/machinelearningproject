```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("water_potability.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
drop = dataset.isnull().sum()/dataset.shape[0]*100
miss = drop[drop>20].keys()
```

```python
drop
```

```python
dataset1 = dataset.drop(miss,axis=1)
dataset1.shape
```

```python
dataset1.isnull().sum()
```

```python
num_cols = dataset1.select_dtypes(["int64","float64"]).keys()
num_cols
```

```python
for i in num_cols:
    dataset1[i].fillna(dataset1[i].mean(),inplace = True) 
```

```python
dataset1.info()
```

```python
X = dataset1.drop('Potability',axis=1)
X
y=dataset1['Potability']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
X_train.shape, X_test.shape, y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier1 = RandomForestClassifier(n_estimators =10 , criterion ='entropy', random_state = 40)
classifier1.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train,y_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
y_pred = classifier1.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel='linear', probability=False, random_state=40)
classifier2.fit(X_train,y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred2,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred2)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier3 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0) 
classifier3.fit(X_train, y_train)
```

```python
y_pred3 = classifier3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred3)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred3))
```

```python
from sklearn.linear_model import LogisticRegression
classifier4 = LogisticRegression(penalty='l2', random_state = 0)
classifier4.fit(X_train,y_train)
```

```python
y_pred4 = classifier4.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred4)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred4,y_test))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred4,y_test)
```
